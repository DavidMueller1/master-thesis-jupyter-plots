```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Data
data_hub = [
    85,
    77.5,
    85,
    87.5,
    75,
    82.5,
    75,
    85,
]

pilot_passenger_data = [1, 0, 0, 1, 1, 0, 0, 1, 0]

# Convert pilot_passenger_data to colors
colors = ['indianred' if user == 1 else 'cornflowerblue' for user in pilot_passenger_data]

patch_list = []
patch_list.append(mpatches.Patch(color='indianred', label='Pilot'))
patch_list.append(mpatches.Patch(color='cornflowerblue', label='Passenger'))
patch_list.append(mpatches.Patch(color='darkorange', label='Average'))

# Create a bar chart showing the distribution of data_hub values
plt.figure(figsize=(5, 3))
plt.bar(range(len(data_hub)), data_hub, color=colors)
plt.ylim(0, 100)
plt.xlabel('Participants')
plt.ylabel('SUS Score')
plt.xticks(range(len(data_hub)), [f'P{(i+1) if i < 1 else i+2}' for i in range(len(data_hub))])
plt.legend(handles=patch_list, loc='lower left')
# Add average line
plt.axhline(y=sum(data_hub) / len(data_hub), color='darkorange', linestyle='--', label='Average')


# Export the plot as PDF
plt.savefig('output/sus_scores_distribution_hub.pdf', format='pdf', bbox_inches='tight')
plt.savefig('output/sus_scores_distribution_hub.png', format='png', bbox_inches='tight')

plt.show()
```
