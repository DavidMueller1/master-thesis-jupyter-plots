# First Piechart Example
This is a work in progress.  I am trying to create a pie chart using matplotlib.

```python
import matplotlib.pyplot as plt

# Data
data_hub = [
    85,
    77.5,
    85,
    87.5,
    75,
    82.5,
    75,
    85,
]

# Create a bar chart showing the distribution of data_hub values
plt.figure(figsize=(10, 6))
plt.bar(range(len(data_hub)), data_hub, color='skyblue')
plt.ylim(0, 100)
plt.xlabel('Participants')
plt.ylabel('SUS Score')
plt.xticks(range(len(data_hub)), [f'P{(i+1) if i < 1 else i+2}' for i in range(len(data_hub))])
# Add average line
plt.axhline(y=sum(data_hub) / len(data_hub), color='red', linestyle='--', label='Average')

# Export the plot as PDF
plt.savefig('output/sus_scores_distribution_hub.pdf', format='pdf')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
data_hub = [
    95,
    82.5,
    90,
    72.5,
    72.5,
    85,
    72.5,
    82.5
]

# Create a bar chart showing the distribution of data_hub values
plt.figure(figsize=(10, 6))
plt.bar(range(len(data_hub)), data_hub, color='skyblue')
plt.ylim(0, 100)
plt.xlabel('Participants', fontsize=16)
plt.ylabel('SUS Score', fontsize=16)
plt.xticks(range(len(data_hub)), [f'P{(i+1) if i < 1 else i+2}' for i in range(len(data_hub))])
# Add average line
plt.axhline(y=sum(data_hub) / len(data_hub), color='red', linestyle='--', label='Average')

# Export the plot as PDF
plt.savefig('output/sus_scores_dashboard.pdf', format='pdf')

plt.show()
```
